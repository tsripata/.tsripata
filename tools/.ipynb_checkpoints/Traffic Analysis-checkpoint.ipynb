```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="darkgrid")
```

## Analyse Traffic Per day

```python
"""
select count(*), datadate from messaging.styxtrafficmessage where datadate >= '20180720'
and datadate <= '20180807'
group by datadate order by datadate desc
"""

dates = ['20180720','20180721','20180722','20180723','20180724','20180725','20180726','20180727','20180728','20180729','20180730','20180731','20180801','20180802','20180803','20180804','20180805','20180806']
requests = [344045456,320780425,311839104,353222729,358695609,345518555,355298779,350586432,321808723,306623580,351660457,356934894,385896726,360001671,354684918,319904220,306339104,362201312]
data = pd.DataFrame(data={'dates': dates, 'requests':requests})
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
sns.lineplot(data=data, x='dates', y='requests')
```

### Hence, the peak is about 385 M request per day.   (Wednesday) --> Pick 20180801 for analysis
